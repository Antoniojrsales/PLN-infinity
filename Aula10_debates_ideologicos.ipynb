```python
#Biblioteca
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
```

```python
#Buscando um arquivo e transformando em um dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/natorjunior/debates-ideologicos/main/dataset.csv')
dataset
```

```python
#Criando um grafico de pizza
dataset.Label.value_counts().plot.pie()
```

```python
#
texto_junto_0 = dataset.query('Label == 0')
texto_junto_1 = dataset.query('Label == 1')
```

```python
texto_junto_0 = ' '.join(texto_junto_0['Text'].values)
texto_junto_1 = ' '.join(texto_junto_1['Text'].values)
```

```python
#Baixar as stop words
nltk.download('stopwords')
# Carregar as stop words em português
stop_words = set(stopwords.words('portuguese'))
# Baixar as stop words
nltk.download('stopwords')
# Lista de palavras (exemplo)
words = texto_junto_0.split()
# Carregar as stop words em português
stop_words = set(stopwords.words('portuguese'))
# Filtrar as palavras, removendo as stop words
filtered_words = [word for word in words if word not in stop_words]

texto_junto_0 = ' '.join(filtered_words)
```

```python
#Baixar as stop words
nltk.download('stopwords')
# Carregar as stop words em português
stop_words = set(stopwords.words('portuguese'))
# Baixar as stop words
nltk.download('stopwords')
# Lista de palavras (exemplo)
words = texto_junto_1.split()
# Carregar as stop words em português
stop_words = set(stopwords.words('portuguese'))
# Filtrar as palavras, removendo as stop words
filtered_words = [word for word in words if word not in stop_words]

texto_junto_1 = ' '.join(filtered_words)
```

```python
text = texto_junto_0
# Gerando a nuvem de palavras
wordcloud = WordCloud(width = 800, height = 800, background_color ='white', min_font_size = 10).generate(text)
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
